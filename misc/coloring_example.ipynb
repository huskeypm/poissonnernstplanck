```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
np.shape(gray_image)
```

#### Save copy of image in each channel. This should maintain a B/W image 

```python
img[:,:,0] = gray_image
img[:,:,1] = gray_image
img[:,:,2] = gray_image
```

```python
imshow(img)
```

```python
# look for red regions
flatred = np.ndarray.flatten(img[:,:,0])
idxr = np.argwhere(flatred>=255)
print np.shape(idxr)

# look for red regions
flatred = np.ndarray.flatten(img[:,:,2])
idxb = np.argwhere(flatred>=255)
print np.shape(idxb)

idx = np.concatenate([idxr,idxb])

# create new image (rasterized)
rawCopy = np.copy(gray_image)

# apply indices
gCopy = np.ndarray.flatten(rawCopy)
gCopy[idx] = 255
gCopy = np.reshape(gCopy, np.shape(rawCopy))

# put into colorized
newImg = np.zeros_like(img)
newImg[:,:,0] = rawCopy
newImg[:,:,1] = gCopy
newImg[:,:,2] = rawCopy


imshow(newImg)
```
