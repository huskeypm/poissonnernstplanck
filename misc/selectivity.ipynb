For the sake of argument, let's say that 
$$P \propto \frac{\phi}{2-\phi} $$
where $\phi$ is the free volume fraction (this is really valid only for cylindrical inclusions), and that we can approximate the effects of a potential of mean force by
$$ P \propto \exp[-E/kT]\frac{\phi}{2-\phi}$$
This latter approximation is probably quite incorrect, but it should at least trend with the data.

Also assume that the unit cell is square with length L and that the pore radii r satisfy 2r < L, s.t. 
$$ \phi = 1 - \pi r^2/L^2$$

```python
L = 1.
eps = 0.01
rs=np.linspace(0.1, L/2. - eps,20)
phis = 1 - np.pi * rs**2 / L**2

# Peff = P/P0 
Peff = phis / (2 - phis)
```

```python
E=1. # [kT]

plt.plot(phis,Peff*np.exp(-E),label="E>0")
plt.plot(phis,Peff, label="E=0")
plt.plot(phis,Peff*np.exp(E),label="E<0")
```

Slopes indicate efficiency in separating ions by size vs. charge 

```python
L =1
def func2(r, pmf):    
    phi = 1 - np.pi * r**2 / L**2  
    Peff = phi/(2-phi)
    Peff = Peff*np.exp(pmf)
    return Peff
    #return x + y
vfunc = np.vectorize(func2)
eps = 0.1
raxis = np.linspace(eps, L/2.-eps, 100)
pmfaxis = np.linspace(-1, 1, 200)
result2 = vfunc(raxis[:,None], pmfaxis[None,:])
```

```python
plt.figure()
f, (ax1,ax2) = plt.subplots(1,2)

im = ax1.pcolormesh(raxis,pmfaxis,result2.T)
f.colorbar(im, ax=ax1)


CS = ax1.contour(raxis, pmfaxis, result2.T, 4, 
                 colors='k',interpolation='none')

#fig, ax1=plt.subplot(1,2,2)
#fig, ax1 = plt.subplots()
rects = ax2.plot(pmfaxis,vfunc(0.1,pmfaxis),'k',label='pmf')
ax2.set_xlabel('PMF')
ax2a = ax2.twiny()
rects2= ax2a.plot(raxis,vfunc(raxis,0),'k--',label='size')
ax2.legend([rects[0],rects2[0]],['pmf','size'])
ax2a.set_xlabel('Channel rad [nm]')
ax2.set_ylabel('Permeation []')
plt.tight_layout()
plt.gcf().savefig("selectivity.png",dpi=300)
```
