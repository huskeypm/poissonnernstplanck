#### Find period 

```python

def GetPeriod(hstrip):    
    # thresh
    #print np.mean(hstrip)
    hstrip[np.where(hstrip - np.mean( hstrip ) < 0 )] = 0

    #plt.axis('equal')
    #pcolormesh(hstrip,cmap="gray")
    sumAxis = np.argmin( np.shape(hstrip)) # find smallest axis
    
    line = np.sum( hstrip, sumAxis)
    #plt.plot(line)#

    # identify peaks (leading edge) 
    thresh = np.array( line > np.mean(line),dtype=float)
    #plt.figure()
    #plt.plot(thresh)
    nPix = np.shape(thresh)[0]
    diff = thresh[1:]-thresh[0:(nPix-1)]
    leadingEdges = np.where(diff>0)[0] 
    totDist = np.array(leadingEdges[-1] - leadingEdges[0],dtype=float)
    nPeaks = np.shape( leadingEdges )[0]
    peakDist = np.int(totDist/(nPeaks-1))

    #plt.figure()
    #plt.plot(thresh)
    return peakDist
```

#### Need to have dark 'bounds' around the high intensity spots for peak finder to work

```python
#hstrip = np.asarray( dst[15:25,:], dtype=float)

kernel = np.ones((3,3),np.float32)/9
smoothed = cv2.filter2D(gray,-1,kernel) 

plt.figure()
myplot(smoothed)

ret,threshed = cv2.threshold(smoothed,150,255,cv2.THRESH_BINARY)
plt.figure()
myplot(threshed)

plt.figure()
# SECOND index is x (in picture), FIRST index is y
xstrip = np.asarray( threshed[65:75,8:83], dtype=float)
myplot(xstrip)
plt.figure()
xperiod = GetPeriod(xstrip)
print xperiod

plt.figure()
# SECOND index is x (in picture), FIRST index is y
ystrip = np.asarray( threshed[0:,45:48], dtype=float)
myplot(ystrip)
plt.figure()
yperiod = GetPeriod(ystrip)
print yperiod




```
