```python
def GmshContours(contours,imgDim, margin=0,
                 fileName="test.geo",
                 refineMesh=True,
                 markers=None,
                 doNotMarkSurfaces=False, # automagically label surfaces 
                 scaleFactor=1.,lc=50.):

    # Make entries for each contour
    lineis = []
    lineNumbersis= []
    lineLoops = []
    pointsEnd=1;linesEnd=1


    plt.plot()
    plt.xlim([0,imgDim[1]])
    plt.ylim([0,imgDim[0]])

    markersDict = dict()
    #[markersDict[i]=[] i for i in markers] 
    for i, marker in enumerate(markers):
     markersDict[marker] = []

    for i,cnt in enumerate(contours):
            #area = cv2.contourArea(cnt)

            # plot
            #plt.figure()
            nPoints = np.shape(cnt)[0]
            print cnt
            print i
            z = np.reshape(cnt,[nPoints,2])
            t = np.arange(nPoints)
            plt.scatter(z[:,0,],z[:,1], c=t)
            #plt.show()

            # convert into gmsh lines 
            linesi,pointsEnd,linesEnd,lineNumbersi,lineLoop = WriteGmshContour(\
                cnt,pointStart=pointsEnd,lineStart=linesEnd,
                scaleFactor=scaleFactor)
            lineis.append(linesi)
            lineNumbersis.append(lineNumbersi)
            lineLoops.append(lineLoop)

            if markers!=None:
              markersDict[markers[i]].append(lineNumbersi)


```
